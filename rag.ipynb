```python
import ollama
def ask_llm(inp):
    rules=f""" you are a financial assistant working at bankx,a banking company,you will need to respond to customers on there quires but only respose to bancking related quires if any otrher quires apart from bancking then response"sorry i could not help with that"respond humanly as possiable.
    The input given is {inp}
    """
    response=ollama.chat(
        model="llama3:8b",
        messages=[{
            "role":"user",
            "content":rules
        }],
        stream=True
    )
    for res in response:
         print(res["message"]["content"],end="",flush=True)
```

```python
while True:
    inp=input("Enter the input")
    ask_llm(inp)
```

```python
from langchain_community.vectorstores.utils import filter_complex_metadata
texts = filter_complex_metadata(texts)
```

```python
v_store.add_documents(texts)
```

```python
v_store._collection.count()
```

```python
v_store._collection.get(ids=["2b5dd5cc-67c9-4c52-bbed-957851b6002d"], include=["embeddings","documents"])
```

```python
retriever = v_store.as_retriever()
```

```python
query = "What is this document about?"

retrieved_docs = retriever.invoke(query)

context = "\n\n".join(
    [doc.page_content for doc in retrieved_docs]
)

print(context)
```

```python
prompt = f"""
Answer the question using only the provided context.

Context:
{context}

Question:
{query}

Answer:
"""
```

```python
print(query)
print(context)
ask_llm(prompt)
```

```python
while True:
    inp = input("Enter input/Query: ")
    retrieved_docs=retriever.invoke(inp)
    context="\n\n".join(
        [doc.page_content for doc in retrieved_docs]
    )
    prompt = f"""
You are a Intelligent RAG assistant and you need to answer quries based on provided context.
If the answer is not present in the provided context then politely return "I donot know"
The input is {inp}
The retrived context is {context}
"""
    ask_llm(prompt)
```
